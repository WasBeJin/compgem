# Строки и форматирование

## Строки как итерируемые объекты

Символы строки можно перебирать в цикле точно также как элементы списка. Рассмотрим несколько примеров такого перебора, комбинируя их с некоторыми функциями и методами строк.

В качестве первого примера распечатаем строку, вставив между символами дефис

```python
string = "Строка текста"
for s in string:
    print(s, end='-')
```

Строку можно преобразовать в список

```python
list_of_signes = list(string)
print(list_of_signes)
```

Которой потом легко можно слить обратно в одну строку

```python
string = ''.join(list_of_signes)
print(string)
```

Как всегда, то, что мы сделали выше с помощью цикла, можно сделать короче, используя метод строки `join`. Обратите что данный метод следует вызывать для того символа, который мы хотим вставить между буквами строки `string`

```python
string = '-'.join(string)
string
```

Мы уже видели, что к элементам строки можно обращаться по индексам. Синтаксис срезов также можно использовать и  со строками, получая доступ к подстрокам одной строки.

```python
# Вычленим из предыдущей строки все дефисы
string[1::2]
```

```python
long_str = ','.join([str(i) for i in range(1,100)])
long_str
```

```python
'-'.join("1,2,3,4".split(','))
```

Разобьем строку на список символов используя в качестве разделителя дефис, а затем соберем обратно (дефис в результате будет удален)

```python
string = ''.join(string.split('-'))
string
```

```python
string = "    123456    "
print(string)
print(string.strip())
```

## Разбор больших кусков текста

Для построения графиков часто необходимо бывает разбить большой кусок текста следующего вида

```python
text = """0.1,1,2
0.2,1,3
0.3,2,3
"""
```

на части. Сделать это можно следующим образом (все функции/методы уже рассматривались в блокноте [Введение в Python](Введение в Python.ipynb)):

```python
(x, y, z) = "123"
print('x =', x, 'y =', y, 'z =', z)
```

```python
l = text.splitlines()
float(l[0].split(',')[0])
```

```python
for (i, string) in enumerate(text.splitlines()):
    (t, x, y) = string.split(',')
    print('Строка номер:', i+1)
    print('t = ', t)
    print('x = ', x)
    print('y = ', y)
```

Обычно при разборе такого куска текста извлекаемые значения лучше собрать в списки:

```python
# Пустые списки
X = []
Y = []
T = []
for string in text.splitlines():
    (t, x, y) = string.split(',')
    T.append(float(t))
    X.append(float(x))
    Y.append(float(y))
print('T =', T, 'X =', X, 'Y =', Y)
```

## Форматирование текста

```python
d = 1/7
print(d)

print("{0:8.6f}".format(d))
print("{0:10.6f}".format(d))
```

```python
print("{0:10.5e}".format(d))
print("{0:10.5g}".format(d))
```

```python
print("{0:+10.6f}".format(d))
print("{0:10.6f}".format(-d))
```

Отформатируем нашу таблицу умножения в ввиде аккуратных столбиков

```python
table = [i * j for j in range(1,10) for i in range(1,10)]
print(table)
```

```python
for i, item in enumerate(table):
    s = '\n' if (i+1) % 9 == 0 else '' 
    print("{0:>3d}".format(item), end=s)
```

Новый стиль форматирования строк введенный в Python 3.6

```python
print(f"Переменная d = {d}")
print(f"Переменная d = {d:5.3f}")
```

```python
l = ['раз', 'два', 'три']
print(f"l[0] = {l[0]}")
```

```python
print("l[0] = {0[0]} l[2] = {0[2]}".format(l))
```
