```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# Подмодуль для примитивов
import matplotlib.patches as mpatches
```

```python
fig01= plt.figure(num=1, figsize=(5, 5), dpi=300)
ax01 = fig01.add_subplot(1, 1, 1)

ax01.grid(visible=True, lw=0.25)
ax01.set(xlim=(-1, 2), ylim=(-1, 2), aspect='equal')

O = np.array([0, 0])
θ1 = 10
θ2 = -80
r = 1.2
a = r*np.array([np.cos(np.deg2rad(θ1)), np.sin(np.deg2rad(θ1))])
b = r*np.array([np.cos(np.deg2rad(θ2)), np.sin(np.deg2rad(θ2))])

pa = mpatches.Polygon(xy=(O, a), fill=False, color='r')
pb = mpatches.Polygon(xy=(O, b), fill=False, color='b')

# Обратить внимание на удвоение радиусов
arc_01 = mpatches.Arc(
    xy=(0, 0),
    width=2*np.linalg.norm(a),
    height=2*np.linalg.norm(b),
    theta1=np.rad2deg(np.arctan2(a[1], a[0])),
    theta2=np.rad2deg(np.arctan2(b[1], b[0])),
    color='gray'
)
ax01.add_patch(arc_01)

ax01.plot(*O, color='k', marker='o')

ax01.add_patch(pa)
ax01.plot(*a, color='r', marker='o', label='a')

ax01.add_patch(pb)
ax01.plot(*b, color='b', marker='o', label='b')

ax01.legend()
```
