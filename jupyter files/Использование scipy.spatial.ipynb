```python
import numpy as np
from scipy.spatial.transform import Rotation

from math import cos, sin, sqrt, pi
```

Из всей библиотеки нам нужен только класс `Rotation` в котором определены методы для создания вращающего объекта. Создать такой объект можно разными способами, а после создания конвертировать из одного представления в другое.

```python
# Создадим вращающий объект из последовательности элементарных матриц
R = Rotation.from_euler(seq='XYZ', angles=(90, 90, 90), degrees=True)
```

```python
# Представим его также последовательностью элементарных матриц, но уже другой
R.as_euler(seq='XYX', degrees=True)
```

```python
# Можно представить и просто в виде матрицы вращения
R.as_matrix()
```

```python
"Кватернион"
R.as_quat()
```

```python
"вектор истинного эйлерова поворота или просто вектор поворота/вращения"
rv = R.as_rotvec() 
"Угол вращения — норма вектора поворота"
θ = np.rad2deg(np.linalg.norm(rv))
"Единичный направляющий вектор оси вращения можно получить, разделив на величину угла"
a = rv / θ
# распечатем значения переменных
print(f"{rv=},\n{θ=},\n{a=}")
```

```python
"Угол вращения"
R.magnitude()
```

```python
# Параметры Родрига–Гамильтона, вычисляются только λ_1, λ_2, λ_3
# λ_0 можно получить из условия нормировки (см. презентацию)"
R.as_mrp()
```

```python
# набор точек, которые хотим вращать
P = np.array([
    (1, 1, 0),
    (0, 1, 0),
    (0, 0, 1)
])
```

```python
# Стандартный способ применить вращение к набору координат
R.apply(P)
```

```python
# альтернативный сапособ применить вращения
R.as_matrix() @ P.T
```

### Пример вращения с помощью кватерниона из презентации

```python
# Поворот на угол π/3
R2 = Rotation.from_quat([0, 0, sin(pi/6), cos(pi/6)])
```

```python
R2.as_euler(seq='XYZ', degrees=True)
```

```python
R2.apply([1, 0, 0])
```

```python
np.rad2deg(R2.magnitude())
```

```python
R2.as_rotvec() / R2.magnitude()
```
