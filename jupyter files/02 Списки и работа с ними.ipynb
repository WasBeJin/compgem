# Списки и работа с ними

## Конструирование списков с помощью списковых сборок

Python позволяет конструировать списки гибким способом, называемым *списковым включением* или *списковой сборкой* (list comprehension).

Начнем рассмотрение с примера, в котором сгенерируем список, содержащий последовательность целых чисел от 1 до 10. Сделаем вначале это с помощью цикла `for`. Создаем пустой список, а затем в цикле заполняем его значениями

```python
l = []

for i in range(1, 11, 1):
    l.append(i)

print(l)
```

То же самое можно сделать короче, если воспользоваться специальным синтаксисом списковой сборки. Фактически мы записываем цикл внутри квадратных скобок.

```python
l = [i for i in range(1, 11, 1)]
print(l)
```

В нашем простом примере можно сконструировать список намного короче, использовав конструктор `list`

```python
l = list(range(1, 11, 1))
print(l)
```

Усложним наш пример, сформировав список квадратов целых чисел. Воспользуемся сперва циклом

```python
l = []

for i in range(1, 11, 1):
    l.append(i**2)

print(l)
```

А теперь сделаем то же самое с помощью списковой сборки

```python
l = [i**2 for i in range(1, 11, 1)]
print(l)
```

То же самое можно сделать и с помощью конструктора `list` но на этот раз запись не будет короче

```python
l = list(i**2 for i in range(1, 11, 1))
print(l)
```

Продолжим усложнять наш пример, вычисляя квадраты только четных чисел от `1` до `10`. Для этого придется использовать условный оператор `if` для выбора только тех элементов которые делятся на `2` без остатка.

```python
l = []

for i in range(1, 11, 1):
    # оператор % позволяет найти остаток от деления
    if i % 2 == 0:
        l.append(i**2)

print(l)
```

То же самое можно сделать с помощью списковой сборки заметно короче

```python
l = [i**2 for i in range(1, 11, 1) if i % 2 == 0]
print(l)
```

Сформируем теперь таблицу умножения воспользовавшись вложенными списками. Сперва сделаем это с помощью цикла, а потом с помощь списковой сборки.

```python
l = []
for i in range(1, 10, 1):
    M = []
    for j in range(1, 10, 1):
        M.append(i * j)
    l.append(M)
l
```

```python
l = [[i * j for j in range(1, 10, 1)] for i in range(1, 10, 1)]
l
```

Можно использовать двойные циклы для формирования одномерного списка

```python
l = []

for i in range(1, 10):
    for j in range(1, 10):
        l.append(i * j)
print(l)
```

```python
l = [i * j for i in range(1,10) for j in range(1,10)]
print(l)
```

```python
# Распечатка в виде таблицы
for (i, item) in enumerate(l):
    if (i+1) % 9 == 0:
        print(item, end='\n')
    else:
        print(item, end=', ')
```

Добавим теперь условие на четность и сформируем таблицу умножения только четных чисел

```python
l = []
for i in range(1, 10, 1):
    if i % 2 == 0:
        M = []
        for j in range(1, 10, 1):
            if j % 2 == 0:
                M.append(i * j)
        l.append(M)
l
```

```python
l = [[i * j for j in range(1, 10, 1) if j % 2 == 0] for i in range(1, 10, 1) if i % 2 == 0]
l
```

Точно также можно опустить внутренние скобки и получить одномерный список

```python
l = [i * j for j in range(1, 10, 1) if j % 2 == 0 for i in range(1, 10, 1) if i % 2 == 0]
print(l)
```

## Использование функции zip()

Параллельный перебор элементов из двух и более списков можно записать короче, если воспользоваться встроенной функцией `zip`, которая последовательно извлекает из переданных ей списков по элементу и формирует их в кортежи.

Пусть у нас есть список координат $X$ и $Y$ из которых надо сформировать список точек, в котором каждая точка представлена парой $(x, y)$ (кортеж). Как и ранее рассмотрим сперва как это сделать в цикле, а затем в списковой сборке.

```python
X = [1, 2, 3, 4, 5, 6]
Y = [1, 2, 3, 4, 5, 6]

Ps = []

for x, y in zip(X, Y):
    # Обратите внимание на двойные скобки
    print(x, y)
    Ps.append((x, y))
    
print(Ps)
```

```python
Ps = [P for P in zip(X, Y)]
print(Ps)

# Иначе
Ps = list(zip(X, Y))
print(Ps)
```

Самостоятельно разберите еще несколько примеров использования списковых сборок совместно с функцией `zip`

```python
l1 = [1, 2, 3]
l2 = ['Один', 'Два', 'Три']

L = []

for (item1, item2) in zip(l1, l2):
    L.append("{0} <=> {1}".format(item1, item2))

print(L)
```

```python
L = ["{0} <=> {1}".format(item1, item2) for (item1, item2) in zip(l1, l2)]
print(L)
```

```python
r = range(0, 10)

L = [l1 + l2 for (l1, l2) in zip(r, r)]
print(L)
```

Важно помнить, что функция `zip()` создает именно *итератор* (а не список) кортежей из переданных ей списков, а значит результатом ее работы можно воспользоваться только один раз. Рассмотрим пример

```python
z = zip(r, r)

print("Тип объекта, возвращаемого функцией zip():", type(z))

print("Проходим по элементам z первый раз")
for item in z:
    print(item, end=' ')

# Это итератор, поэтому после использования в цикле память освободилась!
# Повторим еще раз и убедимся, что в z теперь пусто!
print("\nПроходим по элементам z второй раз")
for item in z:
    print(item)
```

Чтобы из итератора сделать список можно воспользоваться функцией `list()`

```python
z = list(zip(r, r))
print(z)
```

Еще пример

```python
s1 = ['Петя','Вася','Коля']
s2 = [5, 6, 5]
for (x, y) in zip(s1, s2):
    s = "Имя: {0}, возраст: {1}".format(x,y)
    print(s)
```

## Срезы

Списки поддерживают доступ к своим элементам как по одному индексу, так и по диапазону индексов, что позволяет извлекать из списка сразу несколько элементов (т.е. делать *срезы*). Синтаксис срезов в Python  очень похож на такой же функционал языка Фортран, однако надо всегда помнить, что правая граница интервала никогда не входит в диапазон. Рассмотрим ряд примеров.

```python
l = list(range(0,10)) # от 0 до 9
print(l)
```

```python
# Элементы с первого по четвертый
l[0:4]
```

```python
print(l[3:]) # С четвертого и до последнего
print(l[3:-1]) # С четвертого и до последнего
print(l[:4]) # С первого до третьего
print(l[0:-2]) # С первого до предпоследнего
```

```python
# Каждый второй элемент списка с первого по четвертый
print(l[0:5:2])
# Аналогично каждый третий элемент из всего списка
print(l[::3])
```

При формировании среза не создается отдельный объект, а лишь ссылка на соответствующую часть исходного списка, поэтому если осуществить присваивание срезу нового объекта, то исходный список изменится

```python
l[2:4] = [2, 2.5, 3, 3.5]
print(l)
# а вот так нельзя
# l[0:3] = 0
```

## Дополнительные приемы манипуляции элементами списка

```python
l = list(range(-10,1,1))
print(l)

l1 = [item for item in reversed(l)]
print(l1)

l2 = list(reversed(l))
print(l2)
```

Встроенная функция `reversed` возвращает итератор и никак не модифицирует исходный список. Если необходимо изменить исходный список, то можно воспользоваться методом `reversed` для того, чтобы модифицировать его на месте.

```python
l.reverse()
# Исходный список будет изменен
l
```

Элементы списка можно удалять как по индексу, так и по значению. Для удаления элемента по значению, используется метод `remove`. Использовать его нужно осторожно и не злоупотреблять, так как удаление по значению  работает медленно в больших списках и при отсутствии указанного значения, метод выбросит ошибку, что можно привести к падению программы, если вы эту ошибку не перехватите.

```python
l = [1, 2, 3, 4, 1, 1, 1, 'one', 'two']
print(l)
l.remove('one')
print(l)
l.remove(2)
print(l)
```

Для удаления по индексу можно воспользоваться оператором `del`

```python
del l[0]
del l[3]
print(l)
```

Если метод `append` добавляет элемент в конец списка, то метод `pop` наоборот, удаляет последний элемент

```python
l = [1, 2, 3]
l.pop()
print(l)
```
