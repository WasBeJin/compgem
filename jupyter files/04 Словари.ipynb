# Словари и работа с ними

## Создание словаря

Очень удобный тип данных для задания составных структур — словарь (ассоциативный массив, отображение (map)). Представляет собой расширенный вариант массива, где элементы индексируются не просто числами от 0 до N, а ключами общего вида, которые могут иметь любой базовый тип, однако чаще всего используют строки для придания ключам осмысленных названий.

```python
# Пустой словарь
D = {}
print(type(D))
# или используем конструктор
D = dict()
print(type(D))
```

```python
# Для заполнения словаря используем пары key: value
D = {
    'Ключ 1': 'Значение 1',
    'Ключ 2': 'Значение 2',
    'Ключ 3': 'Значение 3'
}
print(D)
```

```python
keys = ['раз', 'два', 'три']
values = [1, 2, 3]
D2 = dict(zip(keys, values))

print(D2)
```

Доступ к элементам словаря осуществляется путем указания ключа

```python
D['Ключ 1']
```

Попытка обратиться к несуществующему ключу не методом `get` вызовет ошибку

```python
D['key 01']
```

Для добавления новых элементов достаточно указать желаемый ключ и присвоить ему соотвествующее значение.

```python
D['Ключ 4'] = [1, 2, 3, 4]
print(D)
```

Операция `+` не работает со славарями, но для объединения двух словарей можно использовать метод `update`

```python
new_D = {
    'Ключ 05': ('a', 'b'),
    'Ключ 1': 'абвгд' 
}
D.update(new_D)
print(D)
```

Важно помнить, что сохраненение порядка ключей в словаре не гарантируется. В нашем примере порядок сохряняется, но при большем количестве элементов он может нарушится. Если порядок ключей важен, то следует воспользоваться типом данных `OrderedDict` из модуля `collections`

## Перебор значений и ключей в цикле

```python
# перебор всех ключей
for key in D:
    print(key)

# перебор всех значений
for value in D.values():
    print(value)

# перебор пар (ключ, значение)
for key, value in D.items():
    print("{0} <=> {1}".format(key, value))
```

Для проверки, существует ли тот или иной ключ в словаре, можно использовать метод `get`

```python
if D.get('Ключ 4') is None:
    print('Ключа № 4 не существует')
else:
    print('Ключ 4 существует')
```

Словари удобно использовать для хранения структуриованных данных: настроек, большого набора параметров и т. д. Также словари используются в функциях для передачи им дополнительных аргументов. Рассмотрим пример использования словаря для хранения данных о студентах.

```python
Student = {
    'Имя' : "Вася",
    'Фамилия' : "Васильевич",
    'Отчество' : "Васильев",
    'Номер студенческого билета' : 123456789,
    'Группа' : 'НК-401'
}
print(Student['Имя'],' из группы ',Student['Группа'])
```

С помощью словаря можно задать структуру данных для хранения параметров, настроек и т.п. Например, можно организовать несколько словарей в виде списка.

```python
Students = [
    {'Имя' : "Вася",
     'Фамилия' : "Васильевич",
     'Отчество' : "Васильев",
     'Номер студенческого билета' : 123456789,
     'Группа' : 'НК-401'},
    {'Имя' : "Петя",
     'Фамилия' : "Петрович",
     'Отчество' : "Петров",
     'Номер студенческого билета' : 123456782,
     'Группа' : 'НП-401'},
    {'Имя' : "Иван",
     'Фамилия' : "Иванович",
     'Отчество' : "Иванов",
     'Номер студенческого билета' : 123456781,
     'Группа' : 'НК-402'}]
Students
```

Ключевая особенность циклов питона — возможность перебора списка из любых элементов: целых/действительных чисел, строк, словарей, каких-то более сложных объектов. Переберем например в цикле модели, заданные выше.

```python
# И теперь переберем его в цикле
for student in Students:
    info = "Студент по имени {0} учится в группе {1}".format(student['Имя'], student['Номер студенческого билета'])
    print(info)
```

```python
names = [student['Имя'] for student in Students if student['Номер студенческого билета'] % 2 != 0]
for name in names:
    print("У ",name," номер студенческого нечетный")
```
