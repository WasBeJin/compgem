# Интерактивные элементы в iPython

В данном блокноте мы рассмотрим небольшой пример, иллюстрирующий ползунок. Для создания ползунка необходимо импортировать метод `interact` из библиотеки `ipywidgets`. Также нам понадобятся библиотеки `numpy` и `matplotlib`

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

Зададим размер изображения, его разрешение и величину шрифта

```python
plt.rcParams['figure.figsize'] = (13.0, 6.0)
plt.rcParams['font.size'] = 20
plt.rcParams['figure.dpi'] = 300
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.linewidth'] = 0.5
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.color'] = 'black'
```

Теперь создадим простую функцию, которая будет рисовать график, зависящий от параметров $a$ и $b$. Эти параметры передаются в качестве аргументов функции.

```python
def replot(a, b):
    '''Функция для рисования простого графика b*sin(a*t)'''

    fig = plt.figure(0)
    ax = fig.add_subplot(1, 1, 1)
    # можно интерактивно менять любые надписи на графике
    ax.set_title("$a = {0:5.2f},b = {1:5.2f}$".format(a,b))
    plot_label = r"${0:5.1f}\cdot\sin({1:5.1f}t)$".format(a, b)
    
    # Рисуем график на отрезке [-4π, 4π]
    t = np.linspace(-4*np.pi, 4*np.pi, 1000)
    y = b*np.sin(a*t)
    
    ax.plot(t, y, lw=1.5, label=plot_label)
    
    ax.legend(ncol=1)
    # обязательно использование show() иначе картинка 
    # не обновляется, а рисуется новая
    plt.show()
```

Воспользуемся методом `interact` для создания ползунка. Первый аргумент метода — это имя функции, которую необходимо запускать для рисования; следующие два аргумента — это диапазон значений двух аргументов нашей функции: `a = (start, stop, step)`

```python
interact(replot, a = (0.0, 10.0, 0.1), b = (0.0, 10.0, 0.1))
```
