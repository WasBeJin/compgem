```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from ipywidgets import interact
import math
import ipywidgets as widgets
```

```python
def task1(a,b,t):
  if b==1:
    f = 1
    b1 = np.arctan(1)
  elif b==2:
    f = 1/2
    b1 = np.arctan(1/(1/2))
  xyz = np.array([[0,0,0,1],
                [2,0,0,1],
                [0,0,2,1],
                [2,0,2,1],
                [1,2,1,1]
                    ])
  faces = [#[2,3,0,1],
          [3,4,2],
          [1,4,3],
          [1,0,4],
          [0,2,4]]
  rad = np.radians(t)
  a1 = np.array([0,2,0]) / 2
  la0 = np.cos(rad/2)
  la = (np.sin(rad/2) * a1)
  xyz1 = xyz[:,:3] - np.array([1,0,1])
  P = xyz1 + ((2*la0)*np.cross(la,xyz1))+2*np.cross(la,np.cross(la,xyz1))
  P+=np.array([1,0,1])
  xyz2 = np.array([[*P[0],1],
                   [*P[1],1],
                   [*P[2],1],
                   [*P[3],1],
                   [*P[4],1]])
  ra = np.radians(a)
  #rb = np.radians(b)
  #f = 1/np.tan(rb)
  M = np.array([[1,0,-f*np.cos(ra),0],
                [0,1,-f*np.sin(b1),0],
                [0,0,0,0],
                [1,0,0,1]])
  #axes= Pxy @ M @ np.eye(4)
  ex = M @ np.array([1,0,0,1])
  ey = M @ np.array([0,1,0,1])
  ez = M @ np.array([0,0,1,1])
  XYZ = xyz2 @ M.T
  XYZ1=XYZ[:,:2]
  #print(ex)
  fig01 = plt.figure(num=1,dpi=200)
  ax01 = fig01.add_subplot(1,1,1)
  #ax01.grid(ls='--',lw=0.5,color='lightgrey')
  ax01.set_aspect('equal')
  ax01.set_xlim(-2, 3)
  ax01.set_ylim(-2, 3)
  #ax01.scatter(*XYZ1[2])
  #print(XYZ1[[0,4,1]][1])
  ax01.axline([0,0],[ex[0],ex[1]],color='r',lw=0.5)
  ax01.axline([0,0],[ey[0],ey[1]],color='g',lw=0.5)
  ax01.axline([0,0],[ez[0],ez[1]],color='b',lw=0.5)
  # ax01.plot([0,10*ex[0][0]],[0,10*ex[1][0]],color='r',lw=0.5)
  # ax01.plot([0,-10*ex[0][0]],[0,-10*ex[1][0]],color='r',lw=0.5)
  # ax01.plot([0,10*ey[0][0]],[0,10*ey[1][0]],color='g',lw=0.5)
  # ax01.plot([0,-10*ey[0][0]],[0,-10*ey[1][0]],color='g',lw=0.5)
  # ax01.plot([0,10*ez[0][0]],[0,10*ez[1][0]],color='b',lw=0.5)
  # ax01.plot([0,-10*ez[0][0]],[0,-10*ez[1][0]],color='b',lw=0.5)
  #ax01.arrow(0,0,ex[0][0],ex[1][0],color="r")
  #ax01.arrow(0,0,*Y1[:2],color="g")
  #ax01.arrow(0,0,*Z1[:2],color="b")
  #ax01.quiver(-3,0,*X1[:2],color="r")
  for i in faces:
    os = 0.5*np.linalg.det(np.array([[XYZ1[i][0][0],XYZ1[i][1][0],XYZ1[i][2][0]],
                                 [XYZ1[i][0][1],XYZ1[i][1][1],XYZ1[i][2][1]],
                                    [1,1,1]]))
    if os>=0 or len(i)==4:
      T = mpatches.Polygon(XYZ1[i],fill=False)
    else:
      T = mpatches.Polygon(XYZ1[i],fill=False,ls='--',lw=0.5)
    ax01.add_patch(T)
w=widgets.Dropdown(options=[('f=1',1),('f=1/2',2)])
interact(task1,a=(0,180,1),b=w,t=(0,360,1))
```
