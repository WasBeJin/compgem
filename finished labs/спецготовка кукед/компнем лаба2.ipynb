```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from ipywidgets import interact
#plt.style.use('black_white.mplstyle')
```

```python
def task1(angle):
    p = np.array([5,4])
    angle_r = np.radians(angle)
    fig01 = plt.figure(dpi=200)
    ax01 = fig01.add_subplot(1,1,1)
    ax01.grid(ls='--',lw=0.5,color='lightgrey')
    ax01.set_aspect('equal')
    #ax01.set(xlim=(-10, 10),ylim=(-10, 10))
    ax01.scatter(p[0],p[1])
    xy=np.array([[-2,-1],[-2,3],[4,3],[3,-3]])
    R = mpatches.Polygon(xy,fill=False)
    rotation = np.array([np.cos(-angle_r),-np.sin(-angle_r),np.sin(-angle_r),np.cos(-angle_r)]).reshape(2,2)
    R1 = mpatches.Polygon(np.array([np.dot(xy[0] - p,rotation)+p,np.dot(xy[1]-p,rotation)+p,np.dot(xy[2]-p,rotation)+p,np.dot(xy[3]-p,rotation)+p]),color= 'g',fill = False)
    C1 = mpatches.Circle(p,np.sqrt((xy[0][0]-p[0])**2+(xy[0][1]-p[1])**2),fill=False,lw= 0.4,ls="--")
    C2 = mpatches.Circle(p,np.sqrt((xy[1][0]-p[0])**2+(xy[1][1]-p[1])**2),fill=False,lw= 0.4,ls="--")
    C3 = mpatches.Circle(p,np.sqrt((xy[2][0]-p[0])**2+(xy[2][1]-p[1])**2),fill=False,lw= 0.4,ls="--")
    C4 = mpatches.Circle(p,np.sqrt((xy[3][0]-p[0])**2+(xy[3][1]-p[1])**2),fill=False,lw= 0.4,ls="--")
    ax01.add_patch(R)
    ax01.add_patch(R1)
    ax01.add_patch(C1)
    ax01.add_patch(C2)
    ax01.add_patch(C3)
    ax01.add_patch(C4)
```

```python
interact(task1,angle=(0,360,1))
```

```python
def task2(angle1,angle2):
    fig02 = plt.figure(num=2, dpi=200)
    ax02 = fig02.add_subplot(1,1,1)
    ax02.set_aspect('equal')
    angle_1r = np.radians(2*angle1)
    angle_2r = np.radians(2*angle2)
    reflection = np.array([np.cos(angle_1r),np.sin(angle_1r),np.sin(angle_1r),-np.cos(angle_1r)]).reshape(2,2)
    reflection2 = np.array([np.cos(angle_2r),np.sin(angle_2r),np.sin(angle_2r),-np.cos(angle_2r)]).reshape(2,2)
    xy = np.array([[-2,-1],[3,2],[4,-3]])
    ax02.axline(np.array([0,0]),slope = np.tan(np.radians(angle1)),color="#299e9e",lw = 0.7,ls="--")
    ax02.axline(np.array([0,0]),slope = np.tan(np.radians(angle2)),color="#5fb95f",lw = 0.7,ls="--")
    ax02.grid(ls='--',lw=0.5,color='lightgrey')
    ax02.set(xlim=(-10, 10),ylim=(-10, 10))
    T = mpatches.Polygon(xy,color ="#FFA500",fill = False)
    T2 = mpatches.Polygon(np.array([np.dot(xy[0],reflection),np.dot(xy[1],reflection),np.dot(xy[2],reflection)]),color = "#00FFFF",fill=False)
    T3 = mpatches.Polygon(np.array([np.dot(np.dot(xy[0],reflection),reflection2),np.dot(np.dot(xy[1],reflection),reflection2),np.dot(np.dot(xy[2],reflection),reflection2)]),color="#1af41a",fill=False)
    ax02.add_patch(T)
    ax02.add_patch(T2)
    ax02.add_patch(T3)
```

```python
interact(task2,angle1=(0,180,1),angle2=(0,180,1))
```

```python
def task3(angle):
    p = 5+4j
    angle_r = np.radians(angle)
    fig01 = plt.figure(num=1,dpi=200)
    ax01 = fig01.add_subplot(1,1,1)
    ax01.grid(ls='--',lw=0.5,color='lightgrey')
    ax01.set_aspect('equal')
    ax01.set(xlim=(-10, 10),ylim=(-10, 10))
    ax01.scatter(np.real(p),np.imag(p))
    xy=np.array([-2-1j,-2+3j,4+3j,3-3j])
    xy1=xy-p
    u = np.exp(angle_r*1j)
    p1 = np.array([np.real(p),np.imag(p)])
    R = mpatches.Polygon(np.array([[np.real(xy[0]),np.imag(xy[0])],[np.real(xy[1]),np.imag(xy[1])],[np.real(xy[2]),np.imag(xy[2])],[np.real(xy[3]),np.imag(xy[3])]]),fill=False)
    R1 =mpatches.Polygon(np.array([[np.real(u*xy1[0])+np.real(p),np.imag(u*xy1[0])+np.imag(p)],[np.real(u*xy1[1])+np.real(p),np.imag(u*xy1[1])+np.imag(p)],[np.real(u*xy1[2])+np.real(p),np.imag(u*xy1[2])+np.imag(p)],[np.real(u*xy1[3])+np.real(p),np.imag(u*xy1[3])+np.imag(p)]]),color= 'g',fill=False)
    #R2 =mpatches.Polygon(np.array([[np.real(np.sqrt(p1[0]**2+p1[1]**2)*u+),np.imag(u*xy[0])],[np.real(u*xy[1]),np.imag(u*xy[1])],[np.real(u*xy[2]),np.imag(u*xy[2])],[np.real(u*xy[3]),np.imag(u*xy[3])]]),color= 'g',fill=False)
    C1 = mpatches.Circle(p1,np.sqrt((np.real(xy[0])-p1[0])**2+(np.imag(xy[0])-p1[1])**2),fill=False,lw= 0.4,ls="--")
    C2 = mpatches.Circle(p1,np.sqrt((np.real(xy[1])-p1[0])**2+(np.imag(xy[1])-p1[1])**2),fill=False,lw= 0.4,ls="--")
    C3 = mpatches.Circle(p1,np.sqrt((np.real(xy[2])-p1[0])**2+(np.imag(xy[2])-p1[1])**2),fill=False,lw= 0.4,ls="--")
    C4 = mpatches.Circle(p1,np.sqrt((np.real(xy[3])-p1[0])**2+(np.imag(xy[3])-p1[1])**2),fill=False,lw= 0.4,ls="--")
    ax01.add_patch(R)
    ax01.add_patch(R1)
    ax01.add_patch(C1)
    ax01.add_patch(C2)
    ax01.add_patch(C3)
    ax01.add_patch(C4)
```

```python
interact(task3,angle=(0,360,1))
```
