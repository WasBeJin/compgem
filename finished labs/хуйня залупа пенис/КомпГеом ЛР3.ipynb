```python
def povorot_vokrug_oy(coord, angle):
    """Вращает фигуру вокруг оси Oy на заданный угол.

    Args:
        coord: Координаты вершин фигуры.
        angle: Угол поворота в градусах.

    Returns:
        Координаты повернутой фигуры.
    """

    # Переводим угол из градусов в радианы
    angle = angle * np.pi/180

    # Извлекаем координаты x, y, z вершин
    xs, ys, zs = zip(*coord)
    
    # Создаем матрицу поворота вокруг оси Oy
    matriza = ([np.cos(angle), 0, -np.sin(angle)], [0, 1, 0], [np.sin(angle), 0, np.cos(angle)])

    # Поворачиваем каждую вершину
    coord_povorot = []    
    for x, y, z in zip(xs, ys, zs):
        # Выполняем умножение координат вершины на матрицу поворота
        coord_povorot.append(np.round((np.array([x, y, z])@matriza).tolist()))   

    # Возвращаем координаты повернутой фигуры
    return np.array(coord_povorot)
```
