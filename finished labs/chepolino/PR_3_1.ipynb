## Часть 2 Визуализация данных

### Вывод графиков целевых полей

```python
# Поле для прогнозирования должно быть выведено в виде линейного графика
# Не забудьте подписать по-своему!
plt.title('contest10', fontdict={'fontsize': 15})
plt.ylabel('contest10')
plt.xlabel('Номер записи')
plt.plot(np.sort(df['contest10'])) #Данные можно отсортировать, чтобы график был более наглядным
plt.show()
```

```python
#Поле для классификации надо отобразить в виде гистограммы
plt.title('Распределение по rank-type', fontdict={'fontsize': 20})
plt.xlabel('rank-type')
plt.ylabel('Количество записей')
#Теперь рисуем распределение - гистограмму, со значением количества опосуммов каждого возраста, а по оси x - все возраста
plt.bar(np.sort(df['rank-type'].unique()),df.groupby(df['rank-type'])['rank-type'].count())
plt.xticks(df['rank-type'].unique(), rotation=45)
plt.tight_layout()
#Отобразим график
plt.show()

```

```python
#Этот график также можно выполнить в виде круговой диаграммы
#Подсчитаем количество опоссумов с каждым возрастом и запишем их в массив
arr = df.groupby(df['rank-type'])['rank-type'].count()
#Создадим массив меток, соответствующих каждому возрасту, и отсортируем их, чтобы порядок меток совпадал с порядком секторов
lab = np.sort(df['rank-type'].unique())
#метод pie строит круговую диаграмму
plt.pie(arr, labels = lab)
plt.show
```

## Дополнительное задание
### Вывод значений остальных полей для каждого класса

```python
#Минимум для пяти полей вывести точечную диаграмму (метод plt.scatter) значений для классов.
#Нужно построить несколько графиков, на каждом одно поле, все значения, для каждого класса своим цветом
#Чтобы узнать число и значения классов используйте:
print(df['rank-type'].unique())
# Если есть класс nan - данный с ним выводить не надо, то есть надо отобразить только возрасты больше нуля


# Точечная диаграмма
contests = ['contest1', 'contest2', 'contest3', 'contest4', 'contest5', 
            'contest6', 'contest7', 'contest8', 'contest9']

# Строим график для каждого конкурса
for CONTEST in contests:
    plt.title(f'Точечная диаграмма для {CONTEST}', fontdict={'fontsize': 15})
    plt.xlabel('Номер записи')
    plt.ylabel(f'ДАННЫЕ ИЗ {CONTEST}')
    
    # Отфильтровываем пустые классы (nan) если есть
    CLEAN_DF = df[df['rank-type'].notna()]
    
    # Для каждого класса - свой цвет
    for CLASS_NAME in CLEAN_DF['rank-type'].unique():
        CLASS_DATA = CLEAN_DF[CLEAN_DF['rank-type'] == CLASS_NAME]
        plt.scatter(CLASS_DATA.index, CLASS_DATA[CONTEST], label=CLASS_NAME)
    
    plt.legend()
    plt.show()
```
